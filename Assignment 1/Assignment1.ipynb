```python
import csv
import pandas as pd
import numpy as np
```

```python
# a) read data from different formats (like csv, xls)

data = pd.read_csv('/content/Heart.csv')
```

```python
# b) Find Shape of Data

shape = data.shape
print(shape)
```

```python
# c) Find Missing Values

data.isnull().sum()
```

```python
# d) Find data type of each column

df = pd.DataFrame(data)
print(df.dtypes)
```

```python
# e) Finding out Zero's
zero_count = (data == 0).sum()
print("Number of Zero's are = ", zero_count)
```

```python
# f) Indexing and selecting data, sort data,

display(data.head())
```

```python
# selecting rows by there index

rows = data.iloc[1:5]
print(rows)
```

```python
# sorting

sorted_data = data.sort_values('Age')
print(sorted_data)
```

```python
# select a single column

first_column = data['Age']
print('First column is selected :')
print(first_column)
```

```python
display(first_column.head(5))
```

```python
# g) Describe attributes of data, checking data types of each column,

data.describe()
```

```python
from enum import unique
# counting unique values of data, format of each column, converting variable  data type (e.g. from long to short, vice versa)

unique_values_Age = data['Age'].unique()
print('Unique values in Ages :')
print(unique_values_Age.sum())

print('Unique values in Sex :')
unique_values_Sex = data['Sex'].unique()
print(unique_values_Sex.sum())

print('Unique values in RestBp :')
unique_values_RestBP = data['RestBP'].unique()
print(unique_values_RestBP.sum())
```

```python
# Format of Each Column

column_format = data.dtypes
print(column_format)
```

```python
# Converting variable datatypes
data['Age'] = data['Age'].astype(float)
print('Datatype of Age = ', data['Age'].dtypes)

data['Sex'] = data['Sex'].astype(float)
print('Datatype of Sex = ', data['Sex'].dtypes)
```
