```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data from CSV

df = pd.read_csv('/content/insurance.csv')
print(df.head())
```

Compute and display summary statistics for each feature available in the dataset. (e.g.minimum value, maximum value, mean, range, standard deviation, variance and percentiles


```python
# Checking Missing values

df.isnull().sum()
```

```python
# Display Maximum and Minimum Values

max_age = df['age'].max()
print('The maximum age is ',max_age)

min_age = df['age'].min()
print('The minimum age is ',min_age)

max_bmi = df['bmi'].max()
print('The maximum bmi is ',max_bmi)

min_bmi = df['bmi'].min()
print('The minimum bmi is ',min_bmi)

max_charges = df['charges'].max()
print('The maximum charges is ', round(max_charges, 2))

min_charges = df['charges'].min()
print('The minimum charges is ', round(min_charges, 2))
```

```python
# Display Mean

mean_age = df['age'].mean()
print('The mean age is ',round(mean_age))

mean_bmi = df['bmi'].mean()
print('The mean bmi is ',round(mean_bmi))

mean_charges = df['charges'].mean()
print('The mean charges is ',round(mean_charges, 2))
```

```python
# Mode

mode_age = df['age'].mode()
print('The mode age is ',mode_age)

mode_bmi = df['bmi'].mode()
print('The mode bmi is ', round(mode_bmi))

mode_charges = df['charges'].mode()
print('The mode charges is ',round(mode_charges, 2))
```

```python
# Range

max_age = max(df['age'])
min_age = min(df['age'])
range_age = max_age - min_age
print('The range of age is ',range_age)

max_bmi = max(df['bmi'])
min_bmi = min(df['bmi'])
range_bmi = max_bmi - min_bmi
print('The range of bmi is ',round(range_bmi))

max_charges = max(df['charges'])
min_charges = min(df['charges'])
range_charges = max_charges - min_charges
print('The range of charges is ',round(range_charges, 2))
```

```python
# Variance

var_age = df['age'].var()
print('The variance of age is ',round(var_age))

var_bmi = df['bmi'].var()
print('The variance of bmi is ',round(var_bmi))

var_charges = df['charges'].var()
print('The variance of charges is ',round(var_charges, 2))
```

```python
# Standard Deviation

std_age = df['age'].std()
print('The standard deviation of age is ',round(std_age))

std_bmi = df['bmi'].std()
print('The standard deviation of bmi is ',round(std_bmi))

std_charges = df['charges'].std()
print('The standard deviation of charges is ',round(std_charges, 2))
```

```python
# Plotting a histogram for each feature in dataset

sns.histplot(df['age'], color='pink', edgecolor='black', bins=20, kde=True, alpha=0.7)
plt.title('Age Distribution', fontsize=16, fontname='monospace')
plt.xlabel('Age', fontsize=12, fontname='monospace')
plt.ylabel('Frequency', fontsize=12, fontname='monospace')
plt.grid(axis='y', linestyle='--')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

sns.histplot(df['bmi'], color='skyblue', edgecolor='black', bins=20, kde=True, alpha=0.7)
plt.title('BMI Distribution', fontsize=16, fontname='monospace')
plt.xlabel('BMI', fontsize=12, fontname='monospace')
plt.ylabel('Frequency', fontsize=12, fontname='monospace')
plt.grid(axis='y', linestyle='--')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

sns.histplot(df['charges'], color='lightgreen', edgecolor='black', bins=20, kde=True, alpha=0.7)
plt.title('Charges Distribution', fontsize=16, fontname='monospace')
plt.xlabel('Charges', fontsize=12, fontname='monospace')
plt.ylabel('Frequency', fontsize=12, fontname='monospace')
plt.grid(axis='y', linestyle='--')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
